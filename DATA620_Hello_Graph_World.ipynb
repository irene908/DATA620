```python
#import packages
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()

l = ('Andre','Beverly','Carol','Diane','Ed','Fernando','Garth','Heather','Ike','Jane') #alphabetic order of the names
G.add_nodes_from(l)

e = (('Andre', 'Beverly'),('Andre', 'Carol'),('Andre', 'Diane'),('Andre', 'Fernando'),
     ('Beverly', 'Diane'),('Beverly', 'Garth'),('Beverly', 'Ed'),
     ('Carol', 'Fernando'),('Carol', 'Diane'),
     ('Diane', 'Fernando'),('Diane', 'Garth'),
     ('Ed', 'Diane'),('Ed', 'Garth'),     
     ('Fernando','Garth'),
     ('Heather', 'Fernando'),('Heather', 'Garth'),('Heather', 'Ike'),
     ('Ike','Jane' ))
G.add_edges_from(e)

pos = {'Andre':(1,4),'Beverly':(1,2),'Carol':(2,5),'Diane':(2,3),'Ed':(2,1),'Fernando':(3,4),'Garth':(3,2),'Heather':(4,3),'Ike':(5,3),'Jane':(6,3)}

nx.draw_networkx_nodes(G,pos=pos,node_size=800,node_color="black", node_shape="s", linewidths=5,alpha=0.2)
nx.draw_networkx_edges(G,pos=pos,edge_color="tab:purple", width=3)
nx.draw_networkx_labels(G,pos=pos,font_size=10, font_color="black")

plt.xlim(0,7)
plt.ylim(0,6)
plt.show()
```

```python
#import the necessary packages
import math
import matplotlib.pyplot as plt
import pandas as pd 
import networkx as nx
from networkx import algorithms as a
```

**Getting the data**

I selected the **Word adjacencies** dataset from http://www-personal.umich.edu/~mejn/netdata/ for this assignment. This dataset is an adjacency network of common adjectives and nouns in the novel **David Copperfield by Charles Dickens**. There are 112 nodes in this dataset and each node has value of either 0 or 1. 

58 Nodes have value as 0 and 54 Nodes have value as 1.

**Load the data**

The dataset is in .gml file so I used read_gml() available in NetworkX to load it.
The number of nodes and edges are printed below.

```python
G= nx.read_gml('adjnoun.gml')
n=G.number_of_nodes()
e=G.number_of_edges()

print(n,e)
```

**View the Graph**

The figure size is set to 10x10 and the Graph is drawn using draw(). 

A title is added as well.

```python
plt.figure(figsize=(10,10))
p=nx.spring_layout(G)
nx.draw(G,p,font_size=12,alpha=1,with_labels=True,node_color='red')
plt.title("Word adjacencies : David Copperfield by Charles Dicken")
plt.show()
```

**Degree Centrality**

For each node the Degree Centrality is calculated using the degree_centrality() available in NetworkX. The calculated result is displayed below.



```python
D=nx.degree_centrality(G)
D
```

**EigenVector Centrality**

For each node the EigenVector Centrality is calculated using the eigenvector_centrality() available in NetworkX. The calculated result is displayed below.

```python
E=nx.eigenvector_centrality(G)
E
```

**Comparison**

For comparison a pandas dataframe df is used. The columns are Degree Centrality and Eigenvector Centrality and the index is the nodes in G.

To quote from : https://www.hackerearth.com/practice/notes/network-analysis-and-centrality-measures-part-i/

**Degree Centrality** : *the node with the highest degree is more important and hence it is having greater impact over rest of the other nodes.*

**Eigenvector Centrality** : *This method assigns a relative value to a node with respect to the nodes it is connected with. So the one which is connected to a node with better importance will always get a better centrality value. For example: If 100 nodes are connected to node A and node A is connected to node B along with 50 other nodes and node C is connected with 60 nodes and not with A, then the node B will get better than node B because node A to which node B is connected is having a higher centrality or degree or importance. This is used in Google Page Ranking algorithm using walking bots method.*

```python
df = pd.DataFrame()
#df.index = list(G)
df['Degree Centrality'] = D.values()
df['Eigenvector Centrality'] = E.values()
df.index = list(G)
df
```
